```python
#task1

class Calculator:
    def __init__(self):
        self.x = 0
        self.y = 0
        print("Let's Calculate!")
    def add(self, x, y):
        self.x += x
        self.y += y
        print('Value 1: ', x)
        print('Operator: ', '+')
        print('Value 2: ', y)
        print('Result: ', x+y)
    def subtract(self, x, y):
        self.x += x
        self.y += y
        print('Value 1: ', x)
        print('Operator: ', '-')
        print('Value 2: ', y)
        if x < y:
            print('Result: ', y-x)
        else:
            print('Result: ', x-y)
    def multiply(self, x, y):
        self.x += x
        self.y += y
        print('Value 1: ', x)
        print('Operator: ', '*')
        print('Value 2: ', y)
        print('Result: ', x*y)
    def devide(self, x, y):
        self.x += x
        self.y += y
        print('Value 1: ', x)
        print('Operator: ', '/')
        print('Value 2: ', y)
        print('Result: ', x/y)
st1 = Calculator()
value1 = int(input())
operator = str(input())
value2 = int(input())
if operator in '+':
    st1.add(value1, value2)
elif operator in '-':
    st1.subtract(value1, value2)
elif operator in '*':
    st1.multiply(value1, value2)
elif operator in '/':
    st1.devide(value1, value2)
```

```python
#task2

class Customer:
    def __init__(self,name):
        self.name=name
    def greet(self,name=None):
        if name!= None:
            print('Hello',self.name+'!')
        else:
            print('Hello!')
    def purchase(self,*food):
        print(self.name + ',','you have purchased',len(food),'item(s):')
        for idx in food:
            print(idx)
            
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#task3

class Panda:
    def __init__(self,name,gender,age):
        self.name=name
        self.gender=gender
        self.age=age
    def sleep(self,hours=""):
        if hours=="":
            return "{}'s duration is unknown thus should have only bamboo leaves".format(self.name)  
        elif int(hours)>=int(3) and int(hours)<=int(5):
            return "{} sleeps {} hours daily and should have Mixed Veggies".format(self.name,hours)
        elif int(hours)>=int(6) and int(hours)<=int(8):
            return "{} sleeps {} hours daily and should have Eggplant & Tofu".format(self.name,hours)
        elif int(hours)>=int(9) and int(hours)<=int(11):
            return "{} sleeps {} hours daily and should have Broccoli Chicken".format(self.name,hours)
panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female", 3)
panda3 = Panda("Ming Ming", "Female", 8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#task4

class Cat:
    def __init__(self,color='white',action='sitting'):
        self.color=color
        self.action=action
    def printCat(self,color="",action=""):
        if color!=None and action!=None:
            print(self.color,"cat is",self.action)
    def changeColor(self,color):
        self.color=color
#===========================================================

c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", " purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()

```

```python
#task5

class Vehicle:
    def __init__(self,x=0,y=0):
        self.x= x
        self.y=y
    def print_position(self):
        print("(" + str(self.x) +"," + str(self.y) + ")")
    def moveUp(self):
        self.y=self.y+1
    def moveDown(self):
        self.y=self.y-1
    def moveRight(self):
        self.x=self.x+1
    def moveLeft(self):
        self.x=self.x-1
        
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#task6

class Programmer:
    def __init__(self,name,programming,experience):
        self.Name = name
        self.Skill = programming
        self.experience = experience
        self.count = 0
    def printDetails(self):
        if self.count == 0:
            print('Horray! A new programmer is born')
        else:
            print('Updating experience of '+self.Name)
        print('Name:',self.Name)
        print('Programming:',self.Skill)
        print('Experience:',self.experience,'years')
    def addExp(self,n):
        self.experience += n
        self.count+=1
p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
#task7

class Student:
    def __init__(self,name,ID,department='CSE'):
        self.Name = name
        self.id = ID
        self.dep = department
    def dailyEffort(self,n):
        self.time = n
    def printDetails(self):
        print('Name:',self.Name)
        print('ID:',self.id)
        print('Department:',self.dep)
        print('Daily Effort:',self.time,'hour(s)')
        if self.time<=2:
            print('Suggestion: Should give more effort!')
        elif self.time<=4:
            print('Suggestion: Keep up the good work!')
        else:
            print('Suggestion: Excellent! Now motivate others.')
harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#task8

class Patient:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def add_Symptom(self,*args):
        for n in range(0,len(args)+1):
            if len(args)==1:
                self.sym = args[0]
            elif len(args)==2:
                self.sym = args[0]+','+args[1]
            else:
                self.sym = args[0] + ',' + args[1]+','+args[2]
    def printPatientDetail(self):
        print('Name:',self.name)
        print('Age:',self.age)
        print('Symptoms:',self.sym)
p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
#task10

class Shinobi:
    def __init__(self,name,rank):
        self.name=name
        self.rank=rank
        self.salary=0
        self.mission=0
        
    def calSalary(self,num):
        self.mission=num
        if self.rank=="Genin":
            self.salary=num*50
        elif self.rank=="Chunin":
            self.salary = num*100
        else:
            self.salary=num*500
    def changeRank(self,rank):
        self.rank=rank
    def printInfo(self):
        print('Name:',self.name,'\nRank:',self.rank,'nNumber of missions:',self.mission,'\nSalary:',self.salary)
naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```
